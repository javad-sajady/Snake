```python
def action(X,Y,Z):
    if(Z=="R"):
        return [X+1,Y]
    if(Z=="L"):
        return [X-1,Y]
    if(Z=="U"):
        return [X,Y+1]
    if(Z=="D"):
        return [X,Y-1]
    


    
 







    
    
class Apple():
    def __init__(self,X,Y):
        self.X=X
        self.Y=Y

    def Location(self):
        return [self.X,self.Y]
                    
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
        
            
            
    
class  Snake():
    def __init__(self,X,Y,Z):
        self.length=2
        A=action(X,Y,Z)
        self.location=[(A[0],A[1],0),(X,Y,1)]
        
    def Head(self):
        head=[]
        for a,b,c in self.location:
            if(c==0 or c==1):
                head.append([a,b])
        return head
    
           
    def Move(self,Z,apple):
        head=self.Head()
        next_move=action(head[0][0],head[0][1],Z)
        A=False
        if(next_move!=head[1]):
            Location=[(next_move[0],next_move[1],0)]
            for a,b,c in self.location:
                if(c!=self.length-1 or next_move==apple):
                    Location.append((a,b,c+1))
            self.location=Location
            if(next_move==apple):
                self.length=self.length+1
                A=True   
        return A
    
     
    def In_Snake(self,A):
        for a,b,c in self.location:
            if(a==A[0] and b==A[1]):
                return True
        return False
    
    
    def Hit(self,X,Y):
        head=self.Head()
        #here we should change if we whant the sanke pass the wall
        if(X==head[0][0] or 0==head[0][0] or Y==head[0][1] or head[0][1]==0 ):
            return True
        
        for a,b,c in self.location:
            if(a==head[0][0] and b==head[0][1] and c!=0):
                return True 

        return False
 

    def Score(self):
        return (self.length-2)*10










    
class Board(Snake,Apple):
    
    def __init__(self,X,Y):
        self.X=X
        self.Y=Y
        self.Over=False
        #snake in the board
        self.snake=Snake(np.random.randint(X-3)+2,np.random.randint(Y-3)+2,np.random.choice(["R","L","D","U"]))
        #apple in the board
        self.apple_maker()
        
    def apple_maker(self):
        A=[np.random.randint(self.X-2)+1,np.random.randint(self.Y-2)+1]
        while(self.snake.In_Snake(A)):
            A=[np.random.randint(self.X-2)+1,np.random.randint(self.Y-2)+1]
        self.apple=Apple(A[0],A[1])
      
    def Board_size(self):
        return self.X,self.Y
    
    
    def Snake_move(self,Z):
        K=False
        A=self.snake.Move(Z,self.apple.Location())
        if(A):
            self.apple_maker()
            K=True
        if(self.snake.Hit(self.X,self.Y)):
            self.Over=True
        return K
                
                
    def Game_Over(self):
        return self.Over
    
    
    def Snake_Location(self):
        A=[]
        for a,b,c in self.snake.location:
            A.append((a,b))
        A=np.array(A)
        return A
    
    def set_snake_location(self,A):
        j=0
        L=[]
        for a,b in A:
            L.append((a,b,j))
            j+=1
        self.snake.location=L
        self.snake.length=len(L)
        
    def set_apple_location(self,A):
        self.apple=Apple(A[0],A[1])
    
    def Apple_Location(self):
        B=self.apple.Location()
        return B
    
    def Score(self):
        return self.snake.Score()
```

```python
import sys
import random
from PIL import Image, ImageTk
from tkinter import Tk, Frame, Canvas, ALL, NW
```

```python
class Cons:
    BOARD_WIDTH = 500
    BOARD_HEIGHT = 500
    DELAY = 1000
    DOT_SIZE = 10
    MAX_RAND_POS = 27
    
     
    
class windows(Board):
    def __init__(self,root=None):
        self.canvas=Canvas(width=Cons.BOARD_WIDTH, height=Cons.BOARD_HEIGHT,
            background="black", highlightthickness=0)
        self.canvas.pack()
        root.title("Snake Game")
        self.A,self.B=0,0
        self.puse=False
        self.new_game_btn=Button(root,text="new game",fg='blue',command =self.new_game)
        self.new_game_btn.place(relx=0.5, rely=0.5, anchor=CENTER)
        self.load_game_btn=Button(root,text="load game",fg='blue',command =self.load_game)
        self.load_game_btn.place(relx=0.5, rely=0.6, anchor=S)
        root.resizable(width=False, height=False)
        self.loadImages()

        
    
        
    def loadImages(self):
        try:
            self.ihead = Image.open("bodys2.png")
            self.head = ImageTk.PhotoImage(self.ihead)
            self.ibody = Image.open("bodys.png")
            self.body = ImageTk.PhotoImage(self.ibody)
            self.iapple = Image.open("bodys3.png")
            self.apple = ImageTk.PhotoImage(self.iapple)
            self.iwall = Image.open("wall.png")
            self.wall = ImageTk.PhotoImage(self.iwall)
        except IOError as e:
            print(e)
            sys.exit(1)
            
        
        
        
    def createObjects(self):
        self.canvas.create_text(30, 10, text="Score: {0}".format(self.board.Score()),
                         tag="score", fill="white")
        self.canvas.create_image(self.board.Snake_Location()[0,0]*22+30, self.board.Snake_Location()[0,1]*22+30, image=self.head, anchor=NW,  tag="head")
        j=0
        for i in self.board.Snake_Location():
            if(j>=1):
                self.canvas.create_image(i[0]*22+30, i[1]*22+30, image=self.body, anchor=NW, tag="body")
            j+=1
        self.canvas.create_image(self.board.Apple_Location()[0]*22+30,self.board.Apple_Location()[1]*22+30, image=self.apple,
            anchor=NW, tag="apple")
        X,Y=self.board.Board_size()
        for i in range(X+1):
            self.canvas.create_image(i*22+30, 0*22+30, image=self.wall, anchor=NW, tag="wall")
            self.canvas.create_image(i*22+30, Y*22+30, image=self.wall, anchor=NW, tag="wall")
             
        for i in range(Y):
            self.canvas.create_image(X*22+30, i*22+30, image=self.wall, anchor=NW, tag="wall")
            self.canvas.create_image(0*22+30, i*22+30, image=self.wall, anchor=NW, tag="wall")  
        self.set_key.destroy()
        self.l_x.destroy()
        self.l_y.destroy()
        self.e_x.destroy()
        self.e_y.destroy()
        self.level.destroy()
        self.level_entry.destroy()
        
        
        
        
     
    def new_game(self):
        self.new_game_btn.destroy()
        self.load_game_btn.destroy()
        self.l_x=Label(root,text="x Dimention of the Game (must be gratter than 4)")
        self.l_x.place(relx=0.5, rely=0.2, anchor=N)
        self.e_x=Entry(root, bd=5)
        self.e_x.place(relx=0.5, rely=0.3, anchor=CENTER)
        self.l_y=Label(root,text="y Dimention of the Game (must be gratter than 4)")
        self.l_y.place(relx=0.5, rely=0.4, anchor=CENTER)
        self.e_y=Entry(root, bd=5)
        self.e_y.place(relx=0.5, rely=0.5, anchor=CENTER)
        self.level=Label(root,text="deficulty level(1 to 10)")
        self.level.place(relx=0.5, rely=0.6, anchor=CENTER)
        self.level_entry=Entry(root, bd=5)
        self.level_entry.place(relx=0.5, rely=0.7, anchor=CENTER)
        self.set_key=Button(root,text="set",fg='blue',command =self.set_func)
        self.set_key.place(relx=0.5, rely=0.8, anchor=CENTER)
        
    
    
    
    
    
    def load_game(self):
        self.new_game_btn.destroy()
        self.load_game_btn.destroy()
        with open("savefile.txt") as f:
            content = f.readlines()
            for x in content:
                K=list(str(x).split(" "))
                self.A=int(K[0])
                self.B=int(K[1])
                self.C=int(K[2])
                L=[]
                for i in range(3,len(K)-2):
                    L.append(int(K[i]))
                L=np.array(L)
                L=L.reshape((int(L.size/2)),2)
                a=int(K[-2])
                b=int(K[-1])
                M=[a,b]
        f.close()
        self.key_move="R"
        self.board=Board(self.A,self.B)
        self.board.set_apple_location(M)
        self.board.set_snake_location(L)
        Cons.DELAY=int((Cons.DELAY/10)*(11-self.C))
        self.update()
        self.canvas.bind_all("<Key>", self.onKeyPressed)
        self.canvas.after(Cons.DELAY, self.onTimer)
                
                
    
    
    def set_func(self):
        self.A=int(self.e_x.get())
        self.B=int(self.e_y.get())
        self.C=int(self.level_entry.get())
        self.e_y.delete(0,"end")
        self.e_x.delete(0,"end")
        self.level_entry.delete(0,"end")
        self.key_move="R"
        Cons.DELAY=int((Cons.DELAY/10)*(11-self.C))
        self.board=Board(self.A,self.B)
        self.createObjects()
        self.canvas.bind_all("<Key>", self.onKeyPressed)
        self.canvas.after(Cons.DELAY, self.onTimer)
        
        
        
        
        
    def onKeyPressed(self, e):
        key = e.keysym
        LEFT_CURSOR_KEY = "Left"
        if key == LEFT_CURSOR_KEY :
            self.key_move="L"
            
        RIGHT_CURSOR_KEY = "Right"
        if key == RIGHT_CURSOR_KEY:
            self.key_move="R"

        RIGHT_CURSOR_KEY = "Up"
        if key == RIGHT_CURSOR_KEY:
            self.key_move="D"

        DOWN_CURSOR_KEY = "Down"
        if key == DOWN_CURSOR_KEY :
            self.key_move="U"
            
        SPACE_CURSOR_KEY="p"
        if key==SPACE_CURSOR_KEY:
            self.puse=True
            
           
        
        
    def onTimer(self):
        self.update()
        X,Y=self.board.Board_size()
        if not self.puse :
            if  not self.board.Game_Over() :
                if(self.board.Score()*self.C<(X-2)*(Y-2)*5):
                    K=self.board.Snake_move(self.key_move)
                    self.update()
                    self.canvas.after(Cons.DELAY, self.onTimer)
                else:
                        self.Victory()
            else:
                self.gameOver()
        else:
            self.Puse()
        
        
        
        
        
        
    def update(self):
        self.canvas.delete(ALL)
        self.canvas.create_text(30, 10, text="Score: {0}".format(self.board.Score()),
                         tag="score", fill="white")
        self.canvas.create_image(self.board.Snake_Location()[0,0]*22+30, self.board.Snake_Location()[0,1]*22+30, image=self.head, anchor=NW,  tag="head")
        j=0
        for i in self.board.Snake_Location():
            if(j>=1):
                self.canvas.create_image(i[0]*22+30, i[1]*22+30, image=self.body, anchor=NW, tag="body")
            j+=1
        self.canvas.create_image(self.board.Apple_Location()[0]*22+30,self.board.Apple_Location()[1]*22+30, image=self.apple,
            anchor=NW, tag="apple")
        X,Y=self.board.Board_size()
        for i in range(X+1):
            self.canvas.create_image(i*22+30, 0*22+30, image=self.wall, anchor=NW, tag="wall")
            self.canvas.create_image(i*22+30, Y*22+30, image=self.wall, anchor=NW, tag="wall")

        for i in range(Y):
            self.canvas.create_image(X*22+30, i*22+30, image=self.wall, anchor=NW, tag="wall")
            self.canvas.create_image(0*22+30, i*22+30, image=self.wall, anchor=NW, tag="wall")
           
        
        
        
        
        
    def Puse(self):
        self.canvas.delete(ALL)
        self.save_game_btn=Button(root,text="Save",fg='blue',command =self.Save)
        self.save_game_btn.place(relx=0.5, rely=0.5, anchor=CENTER)
        self.resume_game_btn=Button(root,text="Resume",fg='blue',command =self.Resume)
        self.resume_game_btn.place(relx=0.5, rely=0.6, anchor=CENTER)
        
        
        
        
    def Save(self):
        self.canvas.delete(ALL)
        file= open("savefile.txt","w+")
        file.write(str(self.A)+" ")
        file.write(str(self.B)+" ")
        file.write(str(self.C)+" ")
        for i,j in self.board.Snake_Location():
            file.write(str(i)+" "+str(j)+" ")
        i=self.board.Apple_Location()
        file.write(str(i[0])+" "+str(i[1]))
        file.close()
        root.destroy()
        
        
        
    def Resume(self):
        self.resume_game_btn.destroy()
        self.save_game_btn.destroy()
        self.puse=False
        print("the game should be resumed")
        self.canvas.bind_all("<Key>", self.onKeyPressed)
        self.canvas.after(Cons.DELAY, self.onTimer)
        
    
    
    
    def gameOver(self):
        self.canvas.delete(ALL)
        self.canvas.create_text(self.canvas.winfo_width() /2, self.canvas.winfo_height()/2,
            text="Game Over at level{} with score {}".format(self.C ,self.board.Score()), fill="white")
    
    
    
    
    def Victory(self):
        self.canvas.delete(ALL)
        if(self.C!=10):
            self.canvas.create_text(self.canvas.winfo_width() /2, self.canvas.winfo_height()/2,
                text="Victory at level {}".format(self.C), fill="white")
            Cons.DELAY=int((Cons.DELAY/10)*(11-self.C))
            self.C+=1
            self.board=Board(self.A,self.B)
            self.createObjects()
            self.canvas.bind_all("<Key>", self.onKeyPressed)
            self.canvas.after(Cons.DELAY, self.onTimer)

        else:
            self.canvas.create_text(self.canvas.winfo_width() /2, self.canvas.winfo_height()/2,
                text="Champion", fill="white")
```
